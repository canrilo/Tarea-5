```python
"""
Metodo que calcula fft de cada señal
Input: 24 vectores
Output: fft de cada vector
"""
#El espacio entre cada senal es 1 minuto
dt=1

#Matriz para las frecuencias de fourier
freq=np.zeros(shape=(400,24),dtype=float) 

#matriz de fft para cada señal (tipo complex)
matrizTransformadas=np.zeros(shape=(400,24),dtype=numpy.complex)

for i in range(N):
    senal=datos[:,i]
    fft_x = fft(senal) / n # FFT Normalized
    matrizTransformadas[:,i]=fft_x
    fft_z=fftfreq(n,dt)
    freq[:,i]=fft_z

```

```python
"""
metodo que plotea la potencia en funcion de la frecuencia
Input: espectros de potencias de cada vector, vector de frecuencias
Output: plotea en logaritmo
En este caso solo esta para una columna
"""
#matriz de valor absoluto
pot=np.zeros((n,N))
for i in range(n):
    for j in range(N):
        pot[i,j]=np.abs(matrizTransformadas[i,j])
        
#print pot

#matriz de cuadrados del valor absoluto
cuadrado=np.zeros((n,N))
for i in range(n):
    for j in range(N):
        cuadrado[i,j]=pot[i,j]**2
    
print len(cuadrado)

#plot de las potencias
for i in range (N):
    potencias=cuadrado[:,i]
    plt.semilogy(potencias)
    title('Espectro de Potencias')
    #loglog
    
```
